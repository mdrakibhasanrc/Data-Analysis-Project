# About Dataset
The World Happiness Report is a landmark survey of the state of global happiness . The report continues to gain global recognition as governments, organizations and civil society increasingly use happiness indicators to inform their policy-making decisions. Leading experts across fields – economics, psychology, survey analysis, national statistics, health, public policy and more – describe how measurements of well-being can be used effectively to assess the progress of nations. The reports review the state of happiness in the world today and show how the new science of happiness explains personal and national variations in happiness.

# Q1: Relationship between Happines and GDP


```python
#plot
plt.figure(figsize=(12,6))
sns.scatterplot(x=data['Ladder score'],y=data['GDP'],hue=data['regional_indicator'])
plt.title('Between Happines score and GDP per capita',fontsize=14)
plt.xlabel('Happines',fontsize=14);
```

# Q9: Asis Region Happines Score,courruption,Healthy life expectancy

```python
south_asia=data[data['regional_indicator']=='South Asia']
east_asia= data[data['regional_indicator']=='East Asia']
southeast_asia=data[data['regional_indicator']=='Southeast Asia']
```

```python
plt.figure(figsize = (16,8))
plt.xticks(rotation=90)
sns.lineplot(x="country_name", y = "Ladder score", data = southeast_asia,marker='o',label='southeast_Asia')
sns.lineplot(x="country_name", y = "Ladder score", data = south_asia,marker='*',label='south_Asia')
sns.lineplot(x="country_name", y = "Ladder score", data = east_asia,marker='*',label='east_Asia')
plt.title('Asia Region Happines Score Trend per country',fontsize=15);
```

```python
plt.figure(figsize = (16,8))
plt.xticks(rotation=90)
sns.lineplot(x="country_name", y = "Perceptions of corruption", data = southeast_asia,marker='o',label='southeast_Asia')
sns.lineplot(x="country_name", y = "Perceptions of corruption", data = south_asia,marker='*',label='south_Asia')
sns.lineplot(x="country_name", y = "Perceptions of corruption", data = east_asia,marker='*',label='east_Asia')
plt.title('Asia Region Perceptions of corruption Trend per country',fontsize=15);
```

```python
plt.figure(figsize = (16,8))
plt.xticks(rotation=90)
sns.lineplot(x="country_name", y = "GDP", data = southeast_asia,marker='o',label='southeast_Asia')
sns.lineplot(x="country_name", y = "GDP", data = south_asia,marker='*',label='south_Asia')
sns.lineplot(x="country_name", y = "GDP", data = east_asia,marker='*',label='east_Asia')
plt.title('Asia Region GDP Trend per country',fontsize=15);
```

# Import Necessary Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Import Dataset

```python
df1=pd.read_csv("zomato.csv",encoding='latin-1')
df2=pd.read_excel('Country-Code.xlsx')
```

```python
#merge Two dataset
df=pd.merge(df1,df2,on='Country Code',how='left')
```

# Data Cleaning

```python
# check duplicated value
df.duplicated().sum()
```

Great! No Duplicated Value

```python
# check null values
df.isna().sum().sort_values(ascending=False)
```

```python
# missing percentage
miss_per=df.isna().sum()/len(df)
miss_per
```

```python
# plot missning value heatmap
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis');
```

```python
# handel Cuisines  columns missing values
df['Cuisines'].mode()
```

```python
df['Cuisines'].fillna(df['Cuisines'].mode(),inplace=True)
```

```python
df.head(2)
```

```python
# drop unwanted columns
df.drop(['Restaurant ID','Country Code'],axis=1,inplace=True)
```

```python
# columns rename
df.rename(columns={'Has Table booking':'Table Booking','Has Online delivery':'Online Delivery','Is delivering now':'Delivering Now'},inplace=True)
```

Great ! Our Data Clean.
